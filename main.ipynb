```python
import ohap
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

df=pd.read_csv('CarEvaluation20221207.csv',encoding="shift-jis")

df=df.drop(0,axis=0)
df=df.astype(float)

y=pd.DataFrame(df.iloc[:,0])
X=pd.DataFrame(df.iloc[:,1:])
st = pd.concat([y, X],axis=1)

y=(y - y.min()) / (y.max() - y.min())

# data Generating
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=40)
X_test
```

```python
import seaborn as sns
sns.heatmap(st.corr(),cmap= sns.color_palette('coolwarm', 10),  annot=True,fmt='.2f', vmin = -1, vmax = 1)
```

```python
from sklearn.neural_network import MLPRegressor #ニューラルネットワーク回帰モデル

clf = MLPRegressor (hidden_layer_sizes=(100,100),random_state=42,max_iter=1000)#デフォルトはsolver="adam"
clf.fit(X_train, y_train)
```

```python
X_test2 = np.array(X_test)

owen = ohap.OHAP()
grouping = [(1,2),(3,4,5),(6,)]
columns = X_test.columns.tolist()
```

```python
import time
start = time.time() # 学習時間計測用

shapvalues = owen.shap(clf, X_test2)#, columns)


elapsed = time.time() - start # 時間計測終了
```

```python
print(f"{elapsed:.14f} s") #shapの計算時間
```

```python
start = time.time() # 学習時間計測用

ohapvalues = owen.ohap(clf, X_test2, grouping)#, columns)


elapsed = time.time() - start # 時間計測終了
```

```python
print(f"{elapsed:.14f} s") #ohapの計算時間
```
